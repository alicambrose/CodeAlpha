# Libraries

```python
from ast import increment_lineno
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

## **Now we will be loading the data**

```python
columns = ['ID','Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width', 'Class_labels']
df = pd.read_csv('Iris.csv', names=columns)
df.head()
```

# **Visualization**

```python
df.describe()
```

```python
sns.pairplot(df,hue= 'Class_labels')
```

# **Separating the input columns and the output columns**

```python
data = df.values

X = data[:,0:5] #To remove last column class_label
Y = data[:,5]   #prints only the last column
```

**Spliting for training and testing**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

# **Support Vector Model**

```python
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(X_train, y_train)
```

```python
prediction1 = model_svc.predict(X_test)

from sklearn.metrics import accuracy_score, confusion_matrix
print(accuracy_score(y_test, prediction1))
print(confusion_matrix(y_test, prediction1))
```

# **Logistic Regression Model**

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)
```

```python
predition2 = model_lr.predict(X_test)

from sklearn.metrics import accuracy_score, confusion_matrix
print(accuracy_score(y_test, predition2)*100)
```

# **Decision Tree Model**

```python
from sklearn.tree import DecisionTreeClassifier
model_dt = DecisionTreeClassifier()
model_dt.fit(X_train, y_train)
```

```python
prediction3 = model_dt.predict(X_test)

from sklearn.metrics import accuracy_score, confusion_matrix
print(accuracy_score(y_test, prediction3)*100)
```

# **Report**

```python
prediction3 = model_dt.predict(X_test)

from sklearn.metrics import classification_report
print(classification_report(y_test, prediction3))
```

# **Manual Test**

```python
X_new = np.array([[1, 3, 2, 1, 0.2], [2, 4.9, 2.2, 3.8, 1.1], [3, 5.3, 2.5, 4.6, 1.9]])
prediction = model_dt.predict(X_new)
print("Prediction of Species: {}".format(prediction))
```
